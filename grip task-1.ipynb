```python
# import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
# Reading the Data 
data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
print("Import successful")
data.head(5)
```

```python
#Visualize the distribution of scores
data.plot(x='Hours', y='Scores', style='*')  
plt.title('Hours vs Percentage', size=30)  
plt.xlabel('Hours studied', size=15)  
plt.ylabel('Percentage obtained', size=15)  
plt.show()
```

**From the above graph we can observe a positive correlation between the hours of study and the scores obtained. 
Plotting a regression line aids the previous statement effectively**

```python
#Regression plot
sns.regplot(x= data['Hours'], y= data['Scores'])
plt.title('Regression Plot of Hours vs Percentage',size=30)
plt.xlabel('Hours studied', size=15)
plt.ylabel('Percentage obtained', size=15)
plt.show()
```

**Now, the line clearly indicates a positve correlation. Hence, more hours spent on studying, increased score.**


**Traning the Model**

```python
#Assign x and y values
x = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values
```

**step 1 - split the data**

```python
train_x, val_x, train_y, val_y = train_test_split(x, y, random_state = 0)
```

**Regression algorithms are used if there is a relationship between the input variable and the output variable. This is apt for our task. Linear regression can be adopted as we observe a linear relationship between the entities.**

```python
#Train the model
reg = LinearRegression()
reg.fit(train_x, train_y)
print("Model trained")
```

**Step 3: Use test data to make predictions**

```python
print(val_x)
val_y = reg.predict(val_x)
```

```python
#Predicted scores
print(val_y)
```

```python
plt.scatter(x=val_x, y=val_y, color="red")
plt.plot(val_x, val_y, color='Black')
plt.title('Actual vs Predicted', size=30)
plt.xlabel('Hours studied', size=15)
plt.ylabel('Percentage obtained', size=15)
plt.show()
```

**Step 4: Evaluate the model**

```python
print("Training Score: ",reg.score(train_x,train_y))
print("Test Score: ",reg.score(val_x,val_y))
```

**As seen above, there is very less difference between Training and Test scores.**


**What will be predicted score if a student studies for 9.25 hrs/ day?**

```python
hr = [9.25]
print("If the student spends " + str(hr) + " hr per day studying", end="")
sol = reg.predict([hr])
print(" he/she will score: "+ str(round(sol[0],2)))
```

**Hence, if the student spends [9.25] hours studying per day he/she will score: 93.89 percent**
